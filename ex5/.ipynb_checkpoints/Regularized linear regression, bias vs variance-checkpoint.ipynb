```python
%matplotlib inline
import scipy.io
import scipy.optimize
import matplotlib.pyplot as plt
import numpy as np

data = scipy.io.loadmat('data/ex5data1')
x_val = data['Xval']
y_val = data['yval']
x_test = data['Xtest']
y_test = data['ytest']
x = np.array(data['X'])
y = np.array(data['y'])
theta = np.ones((2,1)).flatten()
lambda_val = 1
```

## Visualizing the dataset

```python
plt.scatter(x,y,marker='x',c='r')
plt.xlabel('Change in water level (x)')
plt.ylabel('Water flowing out of the dam (y)')
plt.show()
```

## Defining the regularized cost function and gradient

```python
def cost_function(theta,x,y,lambda_val):
    x = np.insert(x,0,1,axis=1)
    theta = theta.reshape(len(x[0]),1)
    h = x.dot(theta)
    m = len(x)
    cost = (1/(2*m))*(((h-y)**2).sum())
    lambda_cost = (lambda_val/(2*m)) * (theta[1]**2).sum()
    return cost+lambda_cost

def gradient(theta,x,y,lambda_val):
    x = np.insert(x,0,1,axis=1)
    theta = theta.reshape(len(x[0]),1)
    h = x.dot(theta)
    m = len(x)
    grad_left = (1/m)* (x.T.dot(h-y))
    grad_right = lambda_val*theta/m
    grad_right[0] = 0  # this is by the rules
    return (grad_left+grad_right).flatten()
```

```python
print(cost_function(theta,x,y,lambda_val))
print(gradient(theta,x,y,lambda_val))
```

## Optimization procedure

```python
def optimize_theta(theta,x,y,lambda_val):
    return scipy.optimize.fmin_cg(cost_function,x0=theta,fprime=gradient,args=(x,y,lambda_val),maxiter=1000,disp=False)
```

```python
result = optimize_theta(theta,x,y,lambda_val)
```

## Visualizing the basic linear model fit

```python
thetas = result.reshape(2,1)
x_copy = x.copy()
x_copy = np.insert(x_copy,0,1,axis=1)
ys = x_copy.dot(thetas)
plt.grid(True)
plt.hold(True)
plt.plot(x_copy[:,1],ys,c='b')
plt.scatter(x,y,marker='x',c='r')
plt.xlabel('Change in water level (x)')
plt.ylabel('Water flowing out of the dam (y)')
plt.show()
```

## Implementing learning curve functionality

```python
def learning_curve(theta,x,y,lambda_val):
    training_error = []
    validation_error = []
    for i in range(1,len(x)+1):
        x_subset = x[0:i,:]
        y_subset = y[0:i,:]
        theta_temp = optimize_theta(theta,x_subset,y_subset,lambda_val)
        training_error.append(cost_function(theta_temp,x_subset,y_subset,0))
        validation_error.append(cost_function(theta_temp,x,y,0))
    return training_error,validation_error
```

### Drawing the behaviour of errors

```python
training_error,validation_error = learning_curve(theta,x,y,lambda_val)

xs = np.arange(1,len(x)+1)
plt.grid(True)
plt.hold(True)
plt.plot(xs,training_error,c='b',label = 'Training error')
plt.plot(xs,validation_error,c='r',label = 'Validation error')
plt.legend()
plt.xlabel('Number of training examples')
plt.ylabel('Error')
plt.show()
```

### Generating polynomial features and normalizing features

```python
#eg if p=3, then we will create x^2 and x^3
def gen_polyfeatures(x,p):
    first_col = x.copy()
    for i in range(p-1):
        power = i+2
        x = np.append(x,first_col**power,axis=1)
    return x


def normalize(x):
    for i in range (len(x[0])):
        col = x[:,i]
        mean=np.mean(col)
        maxx = np.max(col)
        minn = np.min(col)
        rangge = maxx-minn
        for i in range(len(col)):
            col[i] = (col[i]-mean)/rangge
    return x
            
k = gen_polyfeatures(x,3)
print(k[0:4])
f = normalize(k)
print(f[0:4])
```

### Here we will make an 8th degree polynomial regression
    - a interesting parameter is the lambda parameter,
        if we set lambda to 0, then we get a perfect match
        if we set lamba to >1, we get a less of a perfect fit

- obviously the case with lambda = 0 is high variance

```python
#######
lambda_p = 0
polynomial_level = 8
######
x_poly = gen_polyfeatures(x,polynomial_level)
x_poly = normalize(x_poly)
theta_poly = np.ones((len(x_poly[0])+1,1))
result_theta = optimize_theta(theta_poly,x_poly,y,lambda_p)
x_poly = np.insert(x_poly,0,1,axis=1)
ys_poly = x_poly.dot(result_theta)
x_poly = x_poly[:,1:]
def process_data(x,y):
    v=[]
    for i in range(len(x)):
         v.append((x.flatten()[i],ys_poly[i]))
    v = sorted(v, key=lambda x: x[0])
    finalX = []
    finalY = []
    for j in v:
        finalX.append(j[0])
        finalY.append(j[1])
    return finalX,finalY

finX,finY = process_data(x,ys_poly)        
plt.grid(True)
plt.hold(True)
plt.plot(finX,finY,c='b')
plt.scatter(x,y,marker='x',c='r')
plt.xlabel('Change in water level (x)')
plt.ylabel('Water flowing out of the dam (y)')
plt.show()
```

```python
training_error2,validation_error2 = learning_curve(theta_poly,x_poly,y,lambda_p)
xs2 = np.arange(1,len(x)+1)
plt.grid(True)
plt.hold(True)
plt.plot(xs2,training_error2,c='b',label='Training error')
plt.plot(xs2,validation_error2,c='r',label='Validation error')
plt.legend()
plt.xlabel('Number of training examples')
plt.ylabel('Error')
plt.show()
```

#### The obvious explanation for the chart above is: 
- We are dealing with 8th degree polynomial which will surely overfit the data especially with a low lambda
- the lambda which we used is 0, so the variance is really high
- the validation error is initially very high as expected but the error really drops with time

### Now try the same but with lambda = 1

```python
#######
lambda_p = 1
######
x_poly = gen_polyfeatures(x,polynomial_level)
x_poly = normalize(x_poly)
theta_poly = np.ones((len(x_poly[0])+1,1))
result_theta = optimize_theta(theta_poly,x_poly,y,lambda_p)
x_poly = np.insert(x_poly,0,1,axis=1)
ys_poly = x_poly.dot(result_theta)
x_poly = x_poly[:,1:]

finX,finY = process_data(x,ys_poly)        
plt.grid(True)
plt.hold(True)
plt.plot(finX,finY,c='b')
plt.scatter(x,y,marker='x',c='r')
plt.xlabel('Change in water level (x)')
plt.ylabel('Water flowing out of the dam (y)')
plt.show()
```

```python
training_error2,validation_error2 = learning_curve(theta_poly,x_poly,y,lambda_p)
xs2 = np.arange(1,len(x)+1)
plt.grid(True)
plt.hold(True)
plt.plot(xs2,training_error2,c='b',label='Training error')
plt.plot(xs2,validation_error2,c='r',label='Validation error')
plt.legend()
plt.xlabel('Number of training examples')
plt.ylabel('Error')
plt.show()
```
