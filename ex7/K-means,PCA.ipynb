#### 1.2 Finding the closest centroid
        - inputs to the function are: 
             - data - the dataset with 2D points
             - centroids - a matrix with the 2D cluster location
             

```python
def find_closest(data,centroids):
    indices = np.zeros(len(data))
    for i in range(len(data)):
        min_dist = np.inf
        point = data[i]
        for j in range(len(centroids)):
            centroid_pos = centroids[j]
            # here it is important to use absolute value
            dist = np.power(abs((point-centroid_pos)).sum(),2)
            if(dist < min_dist):
                min_dist = dist
                indices[i] = j
    return indices
```

```python
closest_indices = find_closest(data1,[[1,1],[2,2],[4,4],[3,3],[40,40]])
print(closest_indices)
```

#### 1.3 Computing the centroid means
    - inputs are:
        - data - datapoints
        - indices - indexes of the closest centroids

```python
def centroid_mean(data,centroids,indices):
    new_centroids = np.zeros(centroids.shape)
    count_of_assigned = np.zeros(len(centroids),dtype=int)
    for i in indices:
        # here we must transform the indices to int (from float)
        # a stupid but relevant thing
        count_of_assigned[int(i)] += int(1)
    for i in range(len(data)):
        index = int(indices[i])
        new_centroids[index] = np.add(new_centroids[index],data[i])
    for i in range(len(new_centroids)):
        if(count_of_assigned[i] > np.power(10,-5)):
            new_centroids[i] = new_centroids[i]/count_of_assigned[i]
        else:
            new_centroids[i] = centroids[i]
    return new_centroids
```

```python
centroid_mean(data1,np.array([[1,1],[2,2],[4,4],[3,3],[40,40]]),closest_indices)
```

#### 1.4 Random initialization of centroids 
    - we pick N centroid from out initial data set

```python
def init_centroids(n,data):
    centroids = np.zeros((n,data.shape[1]))
    rand_ints = np.random.rand(n,1) * len(data)
    j=0
    for i in rand_ints:
        centroids[j] = data[int(i)]
        j+=1
    return centroids
```

```python
init_centroids(3,data1)
```

#### 1.5 Building the K-Means algorithm

```python
data1  = scipy.io.loadmat('data/ex7data2')['X']
centroids = init_centroids(3,data1)

def k_means(data1,iterations,num_centroids=3):
    centroids = init_centroids(num_centroids,data1)
    for i in range(iterations):
        indices = find_closest(data1,centroids)
        centroids = centroid_mean(data1,centroids,indices)
        ###
    return centroids
```

```python
centroids_best = k_means(data1,10)
print(centroids_best)
```

#### 1.6 Plot the data with clusters marked
    - I'll make a faster and less flexible version with 3 centroid by default

```python
def plot_clusters(data,iterations):
    for i in range(iterations):
        centroids = k_means(data,3)
        indices = find_closest(data,centroids)
        L = [[],[],[]]
        for i in range(len(indices)):
            j = int(indices[i])
            L[j].append(data[i])
        L0 = np.array(L[0])
        L1 = np.array(L[1])
        L2 = np.array(L[2])

        plt.scatter(L0[:,0],L0[:,1],c='r')
        plt.scatter(L1[:,0],L1[:,1],c='b')
        plt.scatter(L2[:,0],L2[:,1],c='g')
        plt.scatter(centroids[:,0],centroids[:,1],marker='x',linewidths=5,c='y')
        plt.show()
```

#### 3.2 Implementing PCA
    - 0. mean normalization + feature scaling
        - mean normalization reduces the mean of dataset to 0
        - feature scaling puts all values within [0,1] range
    - 1. compute covariance matrix
        - sigma = (1/m) * x * x.T
    - 2. use singular value decomposition
        - [U, S, V] = svd(sigma)
        - U has the principal components
        - S is the diagonal matrix used for estimation of variance

#### 3.2.1 Mean normalization + feature scaling
    - 1. mean normalization
        - compute the mean of a value
        - subtract the mean from each value
            - now your mean is 0
    - 2. Feature scaling
        - two approaches: 
            - a) divide all values by the (max-min) value
                - this makes sure values are in [0,1] range
            - b) divide all values by standard deviation
                - this normalizes the data not as much as the first min-max approach

```python
def mean_feature_normalize(data):
    mean = np.mean(data,axis=0)
    data = data - mean
    #min_max_range = np.max(data,axis=0) - np.min(data,axis=0)
    stdev = np.std(data,axis=0)
    data = data/stdev
    return data
```

```python
print(mean_feature_normalize(x)[0:10])
```

#### 3.2.2 Computing the covariance matrix

```python
def compute_sigma(x):
    res = x.T.dot(x)/x.shape[0]
    return res
```

```python
compute_sigma(x)
```

#### 3.2.3 Single value decomposition
    - inputs - the sigma matrix
    - outputs: [U, S, V] =- (read scipy.linalg.svd documentation)

```python
def svd(data):
    x = mean_feature_normalize(data)
    sigma = compute_sigma(x)
    U, S, V = scipy.linalg.svd(sigma,full_matrices = True)
    return U,S,V
```

```python
U,S,V = svd(x)
print ("Principal components are vectors\n",U,'\n')
print("Most important principal component vector is\n",U[:,0])
print()
print ("values of diagonal of matrix S are:\n",S)
```

#### 3.4 Projecting data into lower dimensions
    - after we have the first k principal components,
    we can transform the data into the lower dimensions
    -  matrix U contains principal components,
    we will use onyl the k most important ones
    - z = U[k].T.dot(x)

```python
def project_data(U,k,x):
    reduced_u = U[:,0:k]
    z = x.dot(reduced_u)
    return z
```

### 4. PCA on Face Image dataset
    - loading and showing the images

```python
faces = scipy.io.loadmat('data/ex7faces.mat')['X']
```

```python
import scipy.misc #Used to show matrix as an image
import matplotlib.cm as cm #Used to display images in a specific colormap

def getDatumImg(row):
    """
    Function that is handed a single np array with shape 1x1032,
    crates an image object from it, and returns it
    """
    width, height = 32, 32
    square = row.reshape(width,height)
    return square.T
    
def displayData(myX, mynrows = 10, myncols = 10):
    """
    Function that picks the first 100 rows from X, creates an image from each,
    then stitches them together into a 10x10 grid of images, and shows it.
    """
    width, height = 32, 32
    nrows, ncols = mynrows, myncols

    big_picture = np.zeros((height*nrows,width*ncols))
    
    irow, icol = 0, 0
    for idx in range(nrows*ncols):
        if icol == ncols:
            irow += 1
            icol  = 0
        iimg = getDatumImg(myX[idx])
        big_picture[irow*height:irow*height+iimg.shape[0],icol*width:icol*width+iimg.shape[1]] = iimg
        icol += 1
    fig = plt.figure(figsize=(10,10))
    img = scipy.misc.toimage( big_picture )
    plt.imshow(img,cmap = cm.Greys_r)
```

```python
displayData(faces)
```
