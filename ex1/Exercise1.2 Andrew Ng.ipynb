```python
%matplotlib inline
file = open("/home/dorian/grive/DS learning /Andrew Ng Machine learning/Exercise1/data/ex1data2.txt",'r')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  

x = []
y = []
for line in file.readlines():
    line = line.strip('.').split(',')
    xvalues = [1]
    yfeatures = []
    for i in range (len(line)-1):
        xvalues.append(float(line[i]))
    x.append(xvalues)
    yfeatures.append(float(line[-1]))
    y.append(yfeatures)

# you can only build an numpy array from lists because you can't append data to an array
x = np.array(x)
y = np.array(y)
```

```python
def init_theta():
    theta = []
    temp = np.random.random(len(x[0]))
    theta.append(temp)
    theta = np.array(theta).T
    return theta

def normalize(x):
    for col in range(1,len(x[0])):
        column = x[:,col]
        avg = column.mean()
        diff = column.max() - column.min()
        x[:,col] = [(i-avg)/diff for i in column]
```

```python
m = x.size
alpha = 0.003
normalize(x)
```

```python
'''
there are two options for evaluating the model
-a) we multiply the whole matrix by the theta parameters
    ie. [x00 x10 x20 x30]              [theta1]         [x00*theta1 + x10*theta2 + x20*theta3 + x30*theta4]
        [x01 x11 x21 x31]              [theta2]         [x01*theta1 + x11*theta2 + x21*theta3 + x31*theta4]
        .....                 x        [theta3]     =     ....
        [x0n x1n x2n x3n]              [theta4]         [x0n*theta1 + x1n*theta2 + x2n*theta3 + x3n*theta4]
    ie just x.dot(theta)
-b)
    we multiply each row specially against theta parameters
    - ie extract x[0] ===>  [x00 x10 x20 x30] 
    and with theta being
     [theta1]       
     [theta2]        
     [theta3]  
     [theta4]
    then just x.dot(theta) (ie 1x4 X 4x1 = 1)
    or also we can do theta.T.dot(x.T) by andrew ng notation
'''
def evaluate(theta,x):
    return x.dot(theta)

def costFunction(theta,x,y):
    results = x.dot(theta)
    difference = np.square((y - results))
    cost = difference.sum()/(2*m)
    return cost


def gradient_descent(theta,x,y):
    cost_function_hist = np.zeros(iterations)
    for j in range(iterations):
        # so let's say our theta is 3x1 and our x is 97x3
        # evaluate will return x.dot(theta.T) which is 97x3 X 3x1 = 97x1
        term = (evaluate(theta,x) - y)
        #print(theta)
        theta = theta - alpha/m * x.T.dot(term)
        cost_function_hist[j] = costFunction(theta,x,y)
    return theta,cost_function_hist
```

```python
iterations = 200000
theta = init_theta()

theta,history = gradient_descent(theta,x,y)
plt.plot(history,c='r')
plt.xlabel("Number of iterations")
plt.ylabel("Cost")
plt.ylim()
```
